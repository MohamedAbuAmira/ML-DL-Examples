### Searchable pdf output (with arabic)

```
%%bash
# move the needed file data for any language
# in my case I need arabic data, SO
cp /content/tessdata/ara.traineddata /usr/local/share/tessdata
```

```
%%bash
cd testing
wget https://cloud.githubusercontent.com/assets/17473681/13320324/bc160e22-dbd0-11e5-8090-6f3728fcc06d.jpg
```

```
from PIL import Image
Image.open("/content/testing/bc160e22-dbd0-11e5-8090-6f3728fcc06d.jpg")
```

```
%%bash 
# This creates a pdf with the image and a separate searchable text layer with the recognized text.

tesseract testing/ara_test.jpg testing/test_ara -l ara pdf
```

```
!head -n 100 /content/testing/test_ara.txt
```

```
%%bash
#HOCR output
#Use hocr config file by adding hocr at the end of the command to get the HOCR output.

tesseract testing/eurotext.png testing/eurotext-eng -l eng hocr
```

```
!head -n 100 /content/testing/eurotext-eng.hocr
```

```
%%bash
#TSV output (Currently available in 3.05-dev in master branch on github)
#Use tsv config file by adding tsv at the end of the command to get the TSV output.

tesseract testing/eurotext.png testing/eurotext-eng -l eng tsv
```

```
!head -n 100 /content/testing/eurotext-eng.tsv
```
